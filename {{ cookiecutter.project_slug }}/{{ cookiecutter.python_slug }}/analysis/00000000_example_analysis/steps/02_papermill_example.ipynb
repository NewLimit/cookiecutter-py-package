## Parameters
- `fig_dir`: The directory to save the figures to

```python
from pathlib import Path

import numpy
import pandas
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Parameters
fig_dir = ""
```

```python
Path(fig_dir).mkdir(exist_ok=True, parents=True)

# Make a random plot
data = numpy.random.rand(100, 2) * 100
sns.scatterplot(x=data[:, 0], y=data[:, 1])
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter plot")
plt.grid()
plt.tight_layout()
plt.savefig(Path(fig_dir) / "scatter_plot.png")
plt.show()
plt.close()
```
